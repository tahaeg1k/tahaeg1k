```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def get_file(filename):
    data = []
    f = open(filename)
    #data = f.readline().split(";");
    #data= string_data
    
    for line in f.readlines():
      data.append(float(line[:-1]))
    f.close()
  

    return np.array(data)
```

```python
delta =get_file("drive/MyDrive/M2MO/MonteCarloProject/Filename_Delta.txt")
```

```python
def plot_greek(mc_greek, bs_greek=None, product_name= "European call", greek_name="Delta"):
    plt.figure(figsize=(10, 10))
    plt.subplot(2, 1, 1)

    # Malliavin
    plt.plot(range(len(mc_greek)), mc_greek, label="Malliavin result for "+greek_name, linestyle="--", color='red')

    # Theoretical value
    if bs_greek is not None:
      plt.axhline(y=bs_greek, color='blue', label='Theoretical value in Black-Scholes')

    plt.title(product_name+" "+greek_name)
    plt.xlabel('Number of iterations')
    plt.ylabel(greek_name)
    plt.grid()
    plt.legend()
    
    plt.savefig("Result_"+product_name+"_"+greek_name+".png")
    plt.show()
```

```python
delta =get_file("drive/MyDrive/M2MO/MonteCarloProject/Filename_Delta.txt")
plot_greek(delta[100:], 0.995695, greek_name = "Delta")
```

```python
gamma =get_file("drive/MyDrive/M2MO/MonteCarloProject/Filename_Gamma.txt")
plot_greek(gamma[100:], 0.000421743, greek_name ="Gamma")
```

```python
vega =get_file("drive/MyDrive/M2MO/MonteCarloProject/Filename_Vega.txt")
plot_greek(vega[100:], 1.26523, greek_name ="Vega")
```
